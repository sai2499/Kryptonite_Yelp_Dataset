```python
import psycopg2
  
# establishing the connection
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
  

# Display whole table
cursor.execute("SELECT * FROM yelp_User;")
rows = cursor.fetchall()

user_id = []
year = []

dict_values = {}

for row in rows:
   dict_values[row[0]] = row[8].split(",")
  
# Commit your changes in the database
#conn.commit()
  
# #Closing the connection

```

```python
for key, value in dict_values.items():
    for  i in range(0, len(value)):
        user_id.append(key)
```

```python
for key, value in dict_values.items():
    for i in value:
        if ' ' in i:
            year.append(i.replace(" ",""))
        elif i == 'None' :
            year.append(i)
        else:
            year.append(i)
```

```python
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
create_statement = "CREATE TABLE yelp_elite(user_id varchar(50), year char(8))"
cursor.execute(create_statement)
print("Table created successfully")

conn.commit()
conn.close()
```

```python
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
insert_statement = "INSERT INTO yelp_elite values (%s, %s);"
final_list = []
for i, j in zip(user_id, year):
    final_list.append((i,j))

cursor.executemany(insert_statement, final_list)

conn.commit()
conn.close()
```
